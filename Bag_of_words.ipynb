```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents=[
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are pets."
]
```

```python
vect = CountVectorizer()
```

```python
matrix = vect.fit_transform(documents)
```

```python
print("Vocabulary List:",vect.get_feature_names_out())
```

```python
print("Matrix:\n",matrix.toarray())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
reviews = ["Great product, loved it", "Terrible quality, broke instantly", "Extremely happy with this purchase", "Waste of money"]
labels = [1, 0, 1, 0]  # 1 = Positive, 0 = Negative
```

```python
X_train,X_test,y_train,y_test=train_test_split(reviews,labels,test_size=0.25, random_state=42)
```

```python
vector = CountVectorizer()
X_train_bow=vector.fit_transform(X_train)
X_test_bow=vector.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_bow,y_train)
prediction = model.predict(X_test_bow)
print("Model Accuracy:", accuracy_score(y_test, prediction))
```
